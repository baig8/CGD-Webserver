```python
#Create a function for models
import pickle

def models(X_train, Y_train):
        
    from sklearn.linear_model import LogisticRegression
    #1-Logistic Regression
    log = LogisticRegression(random_state=0)
    log.fit(X_train, Y_train)
    
    # Saving model to disk
    pickle.dump(log, open('model_1.pkl','wb'))

    # Loading model to compare the results
    model = pickle.load(open('model_1.pkl','rb'))
    
    
    
    #2-Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion = 'entropy', random_state=0)
    tree.fit(X_train, Y_train)
    
    pickle.dump(tree, open('model_2.pkl', 'wb'))
    model = pickle.load(open('model_2.pkl', 'rb'))
    
    
    #3-Random Forest Classifier
    
    from sklearn.ensemble import RandomForestClassifier
    
    forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
    forest.fit(X_train, Y_train)
    
    pickle.dump(forest, open('model_3.pkl', 'wb'))
    model = pickle.load(open('model_3.pkl', 'rb'))
    
    #4-Kneighbors Classifiers
    from sklearn.neighbors import KNeighborsClassifier
    knn = KNeighborsClassifier(n_neighbors=7)
    knn.fit(X_train, Y_train)
    
    pickle.dump(knn, open('model_4.pkl', 'wb'))
    model = pickle.load(open('model_4.pkl', 'rb'))
    
    #5-SGD Classifier
    from sklearn.linear_model import SGDClassifier
    sgd = SGDClassifier()
    sgd.fit(X_train, Y_train)
    
    pickle.dump(sgd, open('model_5.pkl', 'wb'))
    model = pickle.load(open('model_5.pkl', 'rb'))
    
    #6-SCV Classifers
    from sklearn.svm import SVC
    s_vc = SVC()
    s_vc.fit(X_train, Y_train)
    
    pickle.dump(s_vc, open('model_6.pkl', 'wb'))
    model = pickle.load(open('model_6.pkl', 'rb'))
    
    #7-GaussianNB
    from sklearn.naive_bayes import GaussianNB
    Gnb = GaussianNB()
    Gnb.fit(X_train, Y_train)
    
    pickle.dump(Gnb, open('model_7.pkl', 'wb'))
    model = pickle.load(open('model_7.pkl', 'rb'))
    
    #8-Bagging Classifiers
    
    from sklearn.ensemble import BaggingClassifier
    BG = BaggingClassifier()
    BG.fit(X_train, Y_train)
    
    pickle.dump(BG, open('model_8.pkl', 'wb'))
    model = pickle.load(open('model_8.pkl', 'rb'))

    
    #9-Ada Boosting Classifier
    
    from sklearn.ensemble import AdaBoostClassifier
    ABC = AdaBoostClassifier()
    ABC.fit(X_train, Y_train)
    
    pickle.dump(ABC, open('model_9.pkl', 'wb'))
    model = pickle.load(open('model_2.pkl', 'rb'))
    
    #10-Gradient Boosting Classifier
    
    from sklearn.ensemble import GradientBoostingClassifier
    GBC = GradientBoostingClassifier()
    GBC.fit(X_train, Y_train)
    
    pickle.dump(GBC, open('model_10.pkl', 'wb'))
    model = pickle.load(open('model_10.pkl', 'rb'))
    
    #11-Extra trees Classifer
    
    from sklearn.ensemble import ExtraTreesClassifier
    ETC = ExtraTreesClassifier()
    ETC.fit(X_train, Y_train)
    
    pickle.dump(ETC, open('model_11.pkl', 'wb'))
    model = pickle.load(open('model_11.pkl', 'rb'))
    
    #12- Linear Discriminant Analysis
    from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
    LDA = LinearDiscriminantAnalysis()
    LDA.fit(X_train, Y_train)
    
    pickle.dump(LDA, open('model_12.pkl', 'wb'))
    model = pickle.load(open('model_12.pkl', 'rb'))
    
    #print('Please choose the model you want to predict')
    #inp = input()
    #for i in range(len(model)):
    
        
      #  while inp <   13:
       #     if inp == 1:
    print('[0]Logistic Regression Training Accuracy:' , log.score(X_train, Y_train))
        #    if inp == 2:
    print('[1]Decision tree Classifier Training ACcuracy:' , tree.score(X_train, Y_train))
          #  if inp == 3:
    print('[2]Random Forest  Classifier Training ACcuracy:' , forest.score(X_train, Y_train))
           # if inp == 4:
    print('[3]KNeighbors  Classifier Training ACcuracy:' , knn.score(X_train, Y_train))
            #if inp == 5:
    print('[4]SGDC  Classifier Training ACcuracy:' , sgd.score(X_train, Y_train))
            #if inp == 6:
    print('[5]SVC  Classifier Training ACcuracy:' , s_vc.score(X_train, Y_train))
            #if inp == 7:
    print('[6]GaussianNB   Classifier Training ACcuracy:' , Gnb.score(X_train, Y_train))
            #if inp == 8:
    print('[7]Bagging   Classifier Training ACcuracy:' , BG.score(X_train, Y_train))
           # if inp ==9:
    print('[8]AdaBoost Classifier Training Accuracy:', ABC.score(X_train, Y_train))
        
            #if inp == 10:
    print('[9] GradientBoosting Classifier Accuracy:', GBC.score(X_train, Y_train))
            #if inp == 11:
    print('[10] ExtraTrees Classsifier ACCuracy: ', ETC.score(X_train, Y_train))
 #           if inp == 12:
    print('[11] Linear Discriminant Analysis:', LDA.score(X_train, Y_train))
#            else:None

    
   
    
    
    
    
    
    
    
    
    
    
    
    return log, tree, forest, knn, sgd, s_vc, Gnb, BG, ABC, GBC, ETC, LDA

    
    
    
    
```

```python
from Bio import SeqIO

count = 0
# we will make an array to save our sequences for nextStep
sequences = []

for seq_record in SeqIO.parse("Brap1Aligned.txt", "fasta"):
    if (count < 6):
        sequences.append(seq_record)
        print("Id: " + seq_record.id + " \t " + "Length: " + str("{:,d}".format(len(seq_record))))
        print(repr(seq_record.seq) + " \n ")
        count = count + 1

# this store the sequence data at 0 in
chr2L = sequences[0].seq

# sequences act like strings in many ways, we can deal with Seq objects as if they were normal python strings , we can do
# operations such as sequence length and iterate over elements

print(len(chr2L))

print("First Letter: " + chr2L[0])
print("Third Letter: " + chr2L[2])
print("Last Letter: " + chr2L[-1])

# counting the number of G in sequence
print("Length:\t" + str(len(chr2L)))
print("G Count:\t" + str(chr2L.count("G")))

# the GC content is important and tells how stable a molecule is
print("GC%:\t\t" + str(100 * float((chr2L.count("G") + chr2L.count("C")) / len(chr2L))))

# the GC content is important and tells how stable a molecule is
print("GC%:\t\t" + str(100 * float((chr2L.count("G") - chr2L.count("C")) /  len(chr2L))))



    ######################################
    # cumulativeSkew(x, seqType):
 
  #          GCSkew = (G - C) / (G + C)
  #          ATSkew = (A - TU) / (A + TU)

   #         sequences.append(GCSkew)
    #        sequences.append(ATSkew)
            ###############################

# -----------slicing a sequence--------
'''when you do a slice the first term is included amd the last is excluded. 2. 
the second thing to notice is that the slice is performed
on the sequence data string but the new object produced is
 another seq object which retains the alphabet information from the orignal seq object. 
 also like a pyton string fgyou can do slices with a start, stop, and stride(the step size which defaults to one)
 for example we can get the first, second and third codon positions of this DNA sequence:'''

chr2LSHORT = chr2L[0:20]
print("Short chr2L:" + chr2LSHORT)
print("Codon pos 1:" + chr2LSHORT[0::3])
print("Codon Pos 2:" + chr2LSHORT[1::3])
print("Codon Pos 3:" + chr2LSHORT[2::3])

# for nucleotide sequences , you can obtain the complement or reverse complement of
# a Seq object using its built-in methods:
print("orignal: " + chr2LSHORT)
print("Complement: " + chr2LSHORT.complement())
print("Reverse Complement : " + chr2LSHORT.reverse_complement())
##---------------Transcription---------------------
''' The actual biological transcription process works from the template strand, doing a reverse
# complement (TCAG --> CUGA) to give the mRNA. However in Biopython and bioinformatics in general,
# we usually work directly with the coding strand because this means we can get the
# mRNA sequence just by switching T --> U
'''
print("Coding DNA: " + chr2LSHORT)
template_dna = chr2LSHORT.reverse_complement()
print("Template DNA: " + template_dna)

###--------REMEMBER by convention nucleotide sequences are normally read from 5' to 3' direction)
# NOW lets transcribe the coding strand into the corresponding mRNA,
# using the Seq object's builtin transcribe method:
messenger_rna = chr2LSHORT.transcribe()
print("Messenger RNA: " + messenger_rna)
# As you can see, all this does is switch T--> U and adjust the alphabet

###--------------Translation---------
# Using a new example lets translate this mRNA into the corresponding protein sequence again taking
# advantage of one of the Seq object' biological method:
print("Protein Sequence: " + messenger_rna.translate())
# or we can directly translate from the coding strand DNA sequence

print("protein sequence: " + chr2LSHORT.translate())


```
